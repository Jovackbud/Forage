```python
transactions.drop(columns=['transaction_id', 'product_id', 'transaction_date', 'online_order', 'order_status', 'product_first_sold_date'], axis=1, inplace=True)
```

```python
transactions.head(3)
```

```python
transactions = transactions.groupby('customer_id')
```

```python
customers = pd.merge(customers, transactions, how='inner', on='customer_id')
```

```python
customers.head(3)
```

## clean customers

```python
customers.drop(columns=['first_name', 'last_name', 'job_title','default','address','country'], axis=1, inplace=True)
```

```python
customers.set_index('customer_id', inplace=True)
```

```python
customers.dropna(inplace=True)
```

```python
customers.head(10)
```

```python
customers['gender'] = customers['gender'].apply(lambda x:x[0].upper())

from datetime import datetime, timedelta
customers ['age'] = (datetime.now() - customers ['DOB']).apply(lambda x: x.days/365.25)

customers ['state'] = customers ['state'].apply(lambda x: x[0].upper())

customers = customers.drop('DOB', axis=1)

```

```python
customers = pd.get_dummies(data= customers, columns=['gender','job_industry_category', 'wealth_segment', 
                                                     'deceased_indicator', 'owns_car', 'state', 'brand','product_line','product_class','product_size',], drop_first=True)
```

```python
customers = customers.groupby(level=0, as_index=True, dropna=True).mean()
```

```python
customers.head(10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = customers.drop(['past_3_years_bike_related_purchases'], axis=1)
y = customers['past_3_years_bike_related_purchases']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(X_train, y_train)
```

```python
pred = dtc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(pred, y_test))
```

```python
print(confusion_matrix(pred, y_test))
```
